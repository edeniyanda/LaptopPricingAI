```python
# Use Group Encoding 
threshold = 9

data["Company"] = data["Company"].apply(lambda x : x if data["Company"].value_counts()[x] >= threshold else "Others")
```

```python
data["Company"].value_counts()
```

```python
# One-hot encoding for Company
CompanyCategories = pd.get_dummies(data['Company'], prefix="Company")

# Join the new columns and drop the original 'Company' column
data = data.join(CompanyCategories).drop("Company", axis=1)
```

```python
data.Product.value_counts()
```

```python
data.TypeName.value_counts()
```

```python
# Drop Product Columns to avoid Noise and High Cardinality 
data = data.drop("Product", axis=1)
```

```python
data
```

```python
# One-hot encoding for TypeName
TypeNameCategories = pd.get_dummies(data.TypeName, prefix="TypeName")

# Join and drop the original column
data = data.join(TypeNameCategories).drop("TypeName", axis=1)
```
