```python
# Use Group Encoding 
threshold = 9

data["Company"] = data["Company"].apply(lambda x : x if data["Company"].value_counts()[x] >= threshold else "Others")
```

```python
data["Company"].value_counts()
```

```python
# One-hot encoding for Company
CompanyCategories = pd.get_dummies(data['Company'], prefix="Company")

# Join the new columns and drop the original 'Company' column
data = data.join(CompanyCategories).drop("Company", axis=1)
```

```python
data.head()
```

```python
data.Product.value_counts()
```

```python
data.TypeName.value_counts()
```

```python
# Drop Product Columns to avoid Noise and High Cardinality 
data = data.drop("Product", axis=1)
```

```python
data
```

```python
# One-hot encoding for TypeName
TypeNameCategories = pd.get_dummies(data.TypeName, prefix="TypeName")

# Join and drop the original column
data = data.join(TypeNameCategories).drop("TypeName", axis=1)
```

```python
data
```

```python
# Viewing the ScreenResolution Columns
data.ScreenResolution.value_counts()
```

```python
# Extract resolution using regex
data['ScreenResolutionWidth'] = data['ScreenResolution'].str.extract(r'(\d+)x')[0].astype(int)
data['ScreenResolutionHeight'] = data['ScreenResolution'].str.extract(r'x(\d+)')[0].astype(int)
```

```python
data
```

```python
# Binary feature for Touchscreen
data['Touchscreen'] = data['ScreenResolution'].apply(lambda x: 1 if 'touchscreen' in x.lower() else 0)
```

```python
# Create a binary feature for 'IPS Panel'
# Checks if the word "IPS Panel" appears in the ScreenResolution column (case-insensitive)
data['IPS_Panel'] = data['ScreenResolution'].apply(lambda x: 1 if 'ips panel' in x.lower() else 0)
```

```python
# Create a binary feature for '4K Display'
# Checks if "4K" appears in the ScreenResolution column (case-insensitive)
data['4K_Display'] = data['ScreenResolution'].apply(lambda x: 1 if '4k' in x.lower() else 0)

# Create a binary feature for 'Retina Display'
# Checks if the word "Retina" appears in the ScreenResolution column (case-insensitive)
data['Retina_Display'] = data['ScreenResolution'].apply(lambda x: 1 if 'retina' in x.lower() else 0)
```
