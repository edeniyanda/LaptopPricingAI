```python
# Function to convert size to MB
def get_to_MB(size):
    if "GB" in size:
        return float(size[:-2]) * 1024  # Convert GB to MB
    elif "TB" in size:
        return float(size[:-2]) * 1024 * 1024  # Convert TB to MB
    return 0

def get_Sum_MB(string):
    if "+" in string:
        string = string.split("+")
        p1, p2 = string[0].strip(), string[1].strip()
        return float(get_to_MB(p1.split(" ")[0])) + float(get_to_MB(p2.split(" ")[0]))
    
```

```python
# Handle expanded storage by splitting and summing up
data['Total_Memory_MB'] = data['Memory'].apply(lambda x : get_Sum_MB(x) if "+" in x else get_to_MB(x.split(" ")[0]))
```

```python
# Function to determine primary storage type
def get_memory_type(memory):
    if 'SSD' in memory:
        return 'SSD'
    elif 'HDD' in memory:
        return 'HDD'
    elif 'Flash Storage' in memory:
        return 'Flash_Storage'
    elif 'Hybrid' in memory:
        return 'Hybrid'
    return 'Unknown'

# Apply the function to determine memory type
data['MemoryType'] = data['Memory'].apply(lambda x: get_memory_type(x))
```

```python
# Function to group operating systems
def group_os(os):
    if 'Windows' in os:
        return 'Windows'
    elif 'macOS' in os or 'Mac' in os:
        return 'Mac'
    elif 'Linux' in os:
        return 'Linux'
    elif 'Chrome' in os:
        return 'Chrome OS'
    elif 'No OS' in os:
        return 'No OS'
    else:
        return 'Other'
```

```python
# Apply the function to group OpSys
data['OpSys_Grouped'] = data['OpSys'].apply(group_os)

# Verify the grouped column
data['OpSys_Grouped'].value_counts()
```

```python
# Apply one-hot encoding
OpSys_Categories = pd.get_dummies(data['OpSys_Grouped'], prefix='OpSys')

# Join the new features and drop the original column
data = data.join(OpSys_Categories).drop(['OpSys', 'OpSys_Grouped'], axis=1)
```
